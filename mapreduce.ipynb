```python
import re


from collections import Counter
my_file = open("/Book1.txt", "r")
  
# reading the file
data = my_file.read()


# Mapping 
data_into_list = data.split("\n ")
newdata =re.split('; |, |\*|\n| ',data)
print(newdata)


```

```python
for word in newdata:
    counts = dict()
    counts[word] = 1 
    print(word+' 1')
```

```python
def word_count():
    counts = dict()


    for word in newdata:
        if word in counts:
            counts[word] += 1
        else: 
            counts[word] = 1 

    return counts

print( word_count())
```
